## Prepare data for table format

### Get the latest rows for percent change

```python
latest_rows = longform_df.tail(2)
latest_rows
```

```python
transposed_latest_rows = latest_rows.transpose()
transposed_latest_rows.head()
```

### Get percentage change

```python
transposed_latest_rows = transposed_latest_rows.dropna()
transposed_latest_rows.head()
```

```python
transposed_latest_rows = transposed_latest_rows.replace('-', pd.NA)
```

```python
transposed_latest_rows['change'] = transposed_latest_rows.apply(
    lambda row: ((float(row.iloc[1]) - float(row.iloc[0])) / float(row.iloc[0]) * 100) if pd.notna(row.iloc[0]) and pd.notna(row.iloc[1]) else None,
    axis=1
)
transposed_latest_rows
```

### Export table

```python
transposed_latest_rows.to_csv('data/bls_table.csv')
```

## Consumer Price Index

```python
data = {
    "seriesid": ['CUSR0000SA0'],  
    "startyear": "2021",
    "endyear": "2026",
    "registrationkey": api_key
}

response = requests.post(url, data=json.dumps(data), headers=headers)
response_data = response.json()
```

```python
response_data
```

```python
series_data = response_data['Results']['series'][0]['data']

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(series_data)
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['date'] = pd.to_datetime(df['year'] + '-' + df['period'].str.replace('M', '') + '-01')
df = df.sort_values('date')

# Calculate 12-month percentage change
df['CPI'] = df['value'].pct_change(periods=12, fill_method=None) * 100

print(df[['date', 'value', 'CPI']].tail())
```

```python
df = df[df['year'] != '2020']
```

```python
df.head()
```

```python
df.to_csv('data/cpi_inflation.csv', index=False)
```
